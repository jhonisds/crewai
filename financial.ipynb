```python
from crewai import Agent, Task, Crew
```

### Agentes
- Pesquisador de Mercado
- Analista de tendÃªncias
- Redator do relatÃ³rio

```python
pesquisador = Agent(
    role="Pesquisador de Mercado",
    goal="Coletar e organizar informaÃ§Ãµes relevantes sobre {sector}",
    backstory="""
    VocÃª Ã© um pesquisador experiente que analisa tendÃªncias de mercado e coleta dados relevantes
    sobre {sector}. Seu trabalho Ã© garantir que todas as informaÃ§Ãµes estejam atualizadas e bem documentadas.   
    """,
    allow_delegation=False,
    verbose=True
)
```

```python
analista = Agent(
    role="Analista de tendÃªncias",
    goal="Analisar os dados {sector} e identificar padrÃµes e oportunidades",
    backstory="""
    VocÃª Ã© um analista de mercado experiente que analisa os dados coletados para identificar
    tendÃªncias emergentes, oportunidades e ameaÃ§as no setor {sector}.    
    """,
    allow_delegation=False,
    verbose=True
)
```

```python
redator = Agent(
    role="Redator do relatÃ³rio",
    goal="Elaborar um relatÃ³rio consolidado sobre a anÃ¡lise de mercado do setor {sector}.",
    backstory="""
    VocÃª Ã© um redator profissional que transforma anÃ¡lises de mercado em um relatÃ³rio estruturado
    e compreensÃ­vel para tomadores de decisÃ£o.     
    """,
    allow_delegation=False,
    verbose=True
)
```

```python
coleta_dados = Task(
    description=(
        "1. Pesquisar e coletar informacÃ§Ãµes atualizadas sobre {sector}"
        "2. Identificar os principais players, tendÃªncias e estatÃ­sticas do setor"
        "3. Organizar os dados de forma clara para anÃ¡lise"        
        ),
    expected_output="Um documento estruturado contendo dados de mercado sobre o {sector}",
    agent=pesquisador
    )
```

```python
analise_tendencias = Task(
    description=(
        "1. Examinar os dados coletados pelo Pesquisador de Mercado."
        "2. Identificar padrÃµes, tendÃªncias emergentes e oportunidades no setor {sector}."
        "3. Elaborar uma anÃ¡lise detalhada destacando os principais pontos"        
        ),
    expected_output="Um relatÃ³rio com insights e tendÃªncias baseados nos dados do setor {sector}",
    agent=analista
    )
```

```python
redacao_relatorio = Task(
    description=(
        "1. Usar a anÃ¡lise de tendÃªncias para criar um relatÃ³rio detalhado sobre o setor {sector}."
        "2. Garantir que o relatÃ³rio esteja bem estruturado e compreensÃ­vel."
        "3. Apresentar um resumo executivo e recomendaÃ§Ãµes finais."        
        ),
    expected_output="Um relatÃ³rio de anÃ¡lise de mercado em formato Markdown e Pdf pronto para leitura e apresentaÃ§Ã£o.",
    agent=redator
    )
```

```python
crew = Crew(
    agents= [pesquisador, analista, redator],
    tasks = [coleta_dados, analise_tendencias, redacao_relatorio],
    verbose=True
    )
```

```python
result = crew.kickoff(inputs={"sector": "AI"})
```
